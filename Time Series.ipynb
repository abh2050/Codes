```python
#time Resampling
df_tesla.head()
```

```python
df_tesla=df_tesla.set_index('Date',drop=True)
```

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()
```

```python
##year end Frequency
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
##Quaterly end frequency
df_tesla.resample(rule='QS').max()
```

```python
df_tesla.resample(rule='QS').max()['Open'].plot()
```

```python
##Business End frequency
df_tesla.resample(rule='BA').max()
```

```python
df_tesla.resample(rule='BA').max()['Open'].plot()
```

```python
df_tesla.resample(rule='BQS').max()['Open'].plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(20,6))
```
